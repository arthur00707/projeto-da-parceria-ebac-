Análise da Crescente Quantidade de Suicídios no Brasil

Nos últimos anos, o Brasil tem enfrentado um aumento expressivo no número de suicídios, refletindo uma tendência preocupante em nossa sociedade. Dados de 2021 revelam que, entre os jovens de 11 a 20 anos, houve um aumento alarmante de 49,6% nos casos de suicídio. No entanto, a maior incidência de mortes por suicídio ocorre na faixa etária de 21 a 30 anos.

Esse crescimento entre os jovens adultos pode ser atribuído a vários fatores, sendo um dos mais significativos a desilusão em relação ao futuro. Muitos jovens enfrentam uma sensação crescente de falta de perspectiva e oportunidades, o que pode levar a um sentimento de desesperança. A pressão para alcançar metas profissionais e pessoais, aliada a um ambiente social em constante mudança e incertezas econômicas, contribui para o aumento do sofrimento emocional nessa faixa etária.

É crucial que abordemos essas questões com sensibilidade e que promovamos o acesso a recursos de saúde mental, além de iniciativas que incentivem o diálogo e o apoio entre os jovens.

Esses exemplos ressaltam a complexidade do fenômeno do suicídio e a necessidade de abordagens diferenciadas para a prevenção em contextos culturais variados.

```python
import matplotlib.pyplot as plt

# Filtrar os dados para o ano de 2021
df_2021 = df[df['Year'] == 2021]

# Lista de continentes ou regiões a serem excluídas
excluded_entities = ['South America', 'North America', 'Africa', 'Asia', 'World', 'Europe']

# Remover as entradas que estão na lista de exclusão
df_2021_filtered = df_2021[~df_2021['Entity'].isin(excluded_entities)]

# Selecionar os 10 países com as maiores taxas de suicídio
top_countries = df_2021_filtered.nlargest(10, 'Total deaths from self-harm amongboth sexes')['Entity'].tolist()

# Filtrar o DataFrame original para incluir apenas esses 10 países ao longo dos anos
df_top_countries = df[df['Entity'].isin(top_countries)]

# Agora, vamos criar o gráfico mostrando a evolução das taxas ao longo do tempo para esses países
plt.figure(figsize=(12, 8))

# Plotar a evolução das taxas de suicídio para cada um dos 10 países
for country in top_countries:
    country_data = df_top_countries[df_top_countries['Entity'] == country]
    plt.plot(country_data['Year'], country_data['Total deaths from self-harm amongboth sexes'], label=country)

# Adicionar título e rótulos
plt.title('Evolução das Taxas de Suicídio para os 10 Países com Maiores Taxas em 2021')
plt.xlabel('Ano')
plt.ylabel('Total de Mortes por Autolesão')
plt.legend(loc='upper left')

# Ajustar o limite do eixo Y
plt.ylim([0, df_top_countries['Total deaths from self-harm amongboth sexes'].max() + 500])

# Adicionar legendas e ajustar a exibição da grade
plt.legend(loc='upper left')
plt.grid(True)

# Ajustar os ticks do eixo X para exibir anos de 3 em 3
years = df_top_countries['Year'].unique()
plt.xticks(ticks=years[::1], rotation=45)  # Exibe a cada 3 anos



# Exibir o gráfico
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Carregar os dados
df_suicide = pd.read_csv("number-of-deaths-from-suicide-ghe.csv")
df_population = pd.read_csv('population-and-demography.csv')  # CSV da população
df_idh = pd.read_csv('human-development-index.csv')

# Filtrar os dados para o ano de 2021 no DataFrame de suicídios
df_suicide_2021 = df_suicide[df_suicide['Year'] == 2021]

# Filtrar os dados para o ano de 2021 no DataFrame de IDH
df_idh_2021 = df_idh[df_idh['Year'] == 2021]

# Mesclar os DataFrames de suicídio e IDH
df_merged = pd.merge(df_suicide_2021, df_idh_2021, on='Entity', suffixes=('_suicide', '_idh'))

# Remover a entrada "World"
df_merged = df_merged[df_merged['Entity'] != 'World']

# Filtrar o DataFrame de população para o ano de 2021
df_population_2021 = df_population[df_population['Year'] == 2021]

# Mesclar os dados de população com os dados de suicídio e IDH
df_merged = pd.merge(df_merged, df_population_2021[['Entity', 'Population - Sex: all - Age: all - Variant: estimates']], on='Entity')

# Calcular a porcentagem de suicídios em relação à população
df_merged['Suicide Rate (%)'] = (df_merged['Total deaths from self-harm amongboth sexes'] / df_merged['Population - Sex: all - Age: all - Variant: estimates']) * 100

# Selecionar os 10 países com a maior taxa de suicídio
top_10_countries = df_merged.nlargest(10, 'Suicide Rate (%)')

# Criar gráfico
fig, ax1 = plt.subplots(figsize=(14, 7))

# Gráfico de barras da taxa de suicídio
sns.barplot(x='Entity', y='Suicide Rate (%)', data=top_10_countries, ax=ax1, color='salmon', label='Taxa de Suicídio (%)', alpha=0.7)
ax1.set_ylabel('Taxa de Suicídio (%)')
ax1.tick_params(axis='y')

# Criar um segundo eixo Y para o IDH
ax2 = ax1.twinx()
sns.lineplot(x='Entity', y='Human Development Index', data=top_10_countries, ax=ax2, color='lightblue', marker='o', label='IDH', linewidth=2)
ax2.set_ylabel('Índice de Desenvolvimento Humano')
ax2.tick_params(axis='y')

# Personalização do gráfico
plt.title('IDH e Taxa de Suicídio em 2021')
ax1.set_xlabel('Países')
ax1.set_xticklabels(top_10_countries['Entity'], rotation=45)
fig.tight_layout()
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')
plt.grid(axis='y')
plt.show()

```
